```python
import nltk
import numpy as np
from keras.utils import np_utils
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout
import sys
import copy
```

```python
dataset = open('data/Fable.txt').read()
dataset = dataset.lower()
```

```python
datas_aswords= (dataset.split())
datas_aswords = datas_aswords
unique_words = sorted(list(set(datas_aswords)))
```

```python
datas_aswords
```

```python
n_words = len(datas_aswords)
n_unique_words = len(unique_words)
n_data = len(dataset)
word_to_int = dict([c, i] for i, c in enumerate(unique_words))
int_to_word = dict([i, c] for i, c in enumerate(unique_words))
```

```python
print(word_to_int)
```

```python
seq_len = 5
dataX = []
dataY = []
for i in range(0, n_words-seq_len, 1):
    seq_in = list(datas_aswords[i:i+seq_len])
    seq_out = datas_aswords[i+seq_len]
    dataX.append(list(word_to_int[in_data] for in_data in seq_in))
    dataY.append(word_to_int[seq_out])
    print("Seq in : ", seq_in, "\nSeq Out : ", seq_out)
```

```python
print(dataX[1:50])
print(dataY[1:50])
```

```python
X = np.reshape(dataX, (n_words-seq_len, seq_len, 1))
print(X.shape)
```

```python
X = X/(n_unique_words)
```

```python
Y = np_utils.to_categorical(dataY)
```

```python
model = Sequential()
model.add(LSTM(n_unique_words*10, input_shape=(seq_len,1)))
model.add(Dropout(0.2))
#model.add(Dropout(0.2))
model.add(Dense(n_unique_words, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X, Y, epochs=6  )
```

```python
data_x = []
data_x = list.copy(dataX[12:13])
print(data_x)
print(len(data_x))
```

```python
for i in range(0, 30):
    x = np.reshape(data_x[0][i:i+seq_len], (1, seq_len, 1))
    #print( x)
    x = x / float(n_unique_words)
    next_word_out = model.predict(x, verbose=2)
   #print("Next Word Out : ", next_word_out)
    next_word_index = np.argmax(next_word_out)
    #print("Max index : ", np.amax(next_word_out, axis=1))
    #print("Next Word index : ", next_word_index)
    next_word = int_to_word[next_word_index]
    #print("Next Word : ", next_word)
    next_word+""
    sys.stdout.write(next_word)
    data_x[0].append(next_word_index)
```
