```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_boston
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
boston = load_boston()
boston.data.shape
boston
```

```python
boston_X = boston.data[:, np.newaxis, 2]
boston_X
```

```python
boston_X_train = boston_X[:-20]
boston_X_test = boston_X[-20:]
```

```python
boston_Y_train = boston.target[:-20]
boston_Y_test = boston.target[-20:]
```

```python
regr = linear_model.LinearRegression()
```

```python
regr.fit(boston_X_train, boston_Y_train)
```

```python
boston_Y_pred = regr.predict(boston_X_test)
```

```python
print("Predicted\tTrue Value")
for i in range(0,len(boston_Y_pred)):
    print("%.2f\t\t%d" % (boston_Y_pred[i], boston_Y_test[i]))
```

```python
print('Coefficients: %.3f' % (regr.coef_))
```

```python
print("Mean Squared error: %.2f" % mean_squared_error(boston_Y_test, boston_Y_pred))
```

```python
print('Variance score: %.2f' % r2_score(boston_Y_test, boston_Y_pred))
```

```python
plt.scatter(boston_X_test, boston_Y_test, color='black')
plt.plot(boston_X_test, boston_Y_pred, color='red', linewidth=3)
plt.xticks(())
plt.yticks(())
plt.show()
```
